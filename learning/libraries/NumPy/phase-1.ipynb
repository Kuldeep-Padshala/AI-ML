## Numpy array and basics
learn about numpy with chai aur code

```python
import numpy as np
```

### creating array from list

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D array: ", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array: ", arr_2d)
```

### List vs numpy array


```python
lis = [1, 2, 3]
print("Python list multiplication", lis * 2)

lis = [1, 2, 3]
print("Python list addition ", lis + [4, 5, 6])

arr = np.array([1, 2, 3])
print("Python array multiplication ", arr * 2)             #element wise multiplication

arr = np.array([1, 2, 3])
print("Python array addition ", arr + [4, 5, 6])             #element wise addition
```

```python
import time
start = time.time()

lis = [i*2 for i in range(1000000)]
print("\n List operation time: ", time.time() - start)

start = time.time()
arr = np.arange(1000000) * 2
print("\n Numpy operation time: ", time.time() - start)
```

### creating array from scratch

```python
zeros = np.zeros((3, 4))
print("\nzeros array: \n", zeros)

ones = np.ones((2, 3))
print("\none array: \n", ones)

full = np.full((2, 2), 7)       # creates a 2x2 array filled with the value 7
print("\nfull array: \n", full)

random = np.random.random((2, 3))
print("\nrandom array: \n", random)

sequence = np.arange(0, 11, 2)    # creates an array with values from 0 to 10 (exclusive) with a step of 2
print("\nsequence array: \n", sequence)
```

### Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3], 
                   [4, 5, 6]])
print("Matrix: ", matrix)

tensor = np.array([[[1, 2], [3, 4]], 
                   [[5, 6], [7, 8]]])
print("Tensor: ", tensor)
```

### Array properties

```python
arr = np.array([[1, 2, 3], 
                [4, 5.6, "abc"]])

print("Shape ", arr.shape)   # returns a tuple representing the dimensions of the array
print("Shape of tensor ", tensor.shape)

print("Dimension ", arr.ndim)  # returns the number of dimensions (axes) of the array
print("Dimension of tensor ", tensor.ndim)

print("Size ", arr.size)  # returns the total number of elements in the array
print("DType ", arr.dtype) # returns the data type of the elements in the array
```

### Array Reshaping

```python
arr = np.arange(12)
print("Original array ", arr)

reshaped = arr.reshape((3, 4))
print("\n Reshaped array ", reshaped)

flattened = reshaped.flatten()
print("\n Flattened array ", flattened)

# ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("\n raveled array ", raveled)

# Transpose
transpose = reshaped.T
print("\n Transposed array ", transpose)
```
