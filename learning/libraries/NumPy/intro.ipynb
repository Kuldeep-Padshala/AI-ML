## Numpy Operations and Functions
Here are some common operations and functions in NumPy along with examples:

1. **Array Creation**:

   `np.array()`: Create an array from a list or tuple.
   - e.g., `arr = np.array([1, 2, 3])`

   `np.zeros()`: Create an array filled with zeros.
   - e.g., `arr = np.zeros((3, 4))`

   `np.ones()`: Create an array filled with ones.
   - e.g., `arr = np.ones((2, 3))`

   `np.arange()`: Create an array with a range of values.
   - e.g., `arr = np.arange(0, 10, 2)`

 - `np.linspace()`: Create an array with evenly spaced values over a specified interval.
   - e.g., `arr = np.linspace(0, 10, 5)`

2. **Array Operations**:

    - Element-wise operations: `+`, `-`, `*`, `/`
    - e.g., `arr1 + arr2`, `arr1 * arr2`

    - Matrix multiplication: `@` or `np.dot()`
    - e.g., `result = arr1 @ arr2` or `result = np.dot(arr1, arr2)`

    - Transpose: `arr.T`
    - e.g., `transposed_arr = arr.T`

    - Reshape: `arr.reshape()`
    - e.g., `reshaped_arr = arr.reshape(2, 3)`
    
    - Flatten: `arr.flatten()`
    - e.g., `flattened_arr = arr.flatten()`

3. **Mathematical Functions**:

    
- `np.sum()`: Sum of array elements.
    - e.g., `total = np.sum(arr)`
    
- `np.mean()`: Mean of array elements.
    - e.g., `average = np.mean(arr)`
    
- `np.std()`: Standard deviation of array elements.
    - e.g., `std_dev = np.std(arr)`
    
- `np.min()`: Minimum value in the array.
    - e.g., `min_value = np.min(arr)`
    
- `np.max()`: Maximum value in the array.
    - e.g., `max_value = np.max(arr)`
    
- `np.argmin()`: Index of the minimum value.
    - e.g., `min_index = np.argmin(arr)`
    
- `np.argmax()`: Index of the maximum value.
    - e.g., `max_index = np.argmax(arr)`

4. **Logical Operations**:

    
- `np.where()`: Return elements based on a condition.
    - e.g., `result = np.where(arr > 2, 'Greater', 'Lesser')`
    
- `np.logical_and()`: Element-wise logical AND.
    - e.g., `result = np.logical_and(arr1 > 0, arr2 < 5)`
    
- `np.logical_or()`: Element-wise logical OR.
    - e.g., `result = np.logical_or(arr1 > 0, arr2 < 5)`

5. **Statistical Functions**:
    
- `np.corrcoef()`: Correlation coefficient.
    - e.g., `correlation = np.corrcoef(arr1, arr2)`
    
- `np.histogram()`: Compute the histogram of a dataset.
    - e.g., `hist, bin_edges = np.histogram(arr, bins=5)`

6. **Linear Algebra Functions**:
    
- `np.linalg.inv()`: Compute the inverse of a matrix.
    - e.g., `inverse = np.linalg.inv(matrix)`
    
- `np.linalg.det()`: Compute the determinant of a matrix.
    - e.g., `determinant = np.linalg.det(matrix)`
    
- `np.linalg.eig()`: Compute the eigenvalues and right eigenvectors of a square array.
    - e.g., `eigenvalues, eigenvectors = np.linalg.eig(matrix)

7. **Random Number Generation**:
    
- `np.random.rand()`: Generate random numbers in the range [0, 1
    - e.g., `random_numbers = np.random.rand(5)`
    
- `np.random.randint()`: Generate random integers.
    - e.g., `random_integers = np.random.randint(0, 10, size=5)`
    
- `np.random.normal()`: Generate random numbers from a normal distribution.
    - e.g., `random_numbers = np.random.normal(0, 1, size=5)`

8. **Array Manipulation**:
    
- `np.concatenate()`: Join arrays along an axis.
    - e.g., `result = np.concatenate((arr1, arr2), axis=0)`
    
- `np.split()`: Split an array into multiple sub-arrays.
    - e.g., `sub_arrays = np.split(arr, 3)`
    
- `np.unique()`: Find unique elements in an array.
    - e.g., `unique_elements = np.unique(arr)`
