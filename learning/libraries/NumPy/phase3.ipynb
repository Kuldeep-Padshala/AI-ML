## Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt # type: ignore
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 120000, 140000, 160000, 190000],  # Starbucks
    [2, 150000, 180000, 220000, 250000],  # RV's Kitchen
    [3, 200000, 230000, 260000, 300000],  # McDonald's
    [4, 180000, 210000, 240000, 270000],  # Dominos Pizza
    [5, 160000, 185000, 205000, 230000]   # La pinoz Pizza
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restau: \n", sales_data[:3])
```

```python
# total sales per year
print("\n Total sales per year: ")
print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)     # here in [:, 1:], : means all rows and 1: means from column index 1 to end
print(yearly_total)

# Minimum sales per restaurant
print("\n Minimum sales per restaurant: ")
min_sales = np.min(sales_data[:, 1:], axis=1)     
print(min_sales)

# Maximum sales per year
print("\n Maximum sales per year: ")
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# Average sales per restaurant
print("\n Average sales per restaurant: ")
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(sales_data[0, 1:], marker='o', label='Starbucks')
plt.plot(sales_data[1, 1:], marker='s', label='RV\'s Kitchen')
plt.plot(sales_data[2, 1:], marker='^', label='McDonald\'s')
plt.plot(sales_data[3, 1:], marker='d', label='Dominos Pizza')
plt.plot(sales_data[4, 1:], marker='v', label='La pinoz Pizza')
plt.title("Average cumulative sales across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Graph to show sales of each restaurant on each year, not same as above line chart which shows cumulative sales of each restaurant across years. This graph will show sales of each restaurant for each year separately.
# on each restaurants(x-axis), there should be 4 bars for 4 years, and height of each bar will represent sales of that restaurant in that year. We can use different colors for bars of different years to differentiate them visually. We can also add legend to indicate which color corresponds to which year. This graph will help us compare sales of different restaurants in each year and identify trends or patterns in their performance over time.

plt.figure(figsize=(10, 6))
bar_width = 0.15

index = np.arange(sales_data.shape[0])  # number of restaurants

plt.bar(index, sales_data[:, 1], width=bar_width, label='2021')
plt.bar(index + bar_width, sales_data[:, 2], width=bar_width, label='2022')
plt.bar(index + 2 * bar_width, sales_data[:, 3], width=bar_width, label='2023')
plt.bar(index + 3 * bar_width, sales_data[:, 4], width=bar_width, label='2024')

plt.xlabel('Restaurants')
plt.ylabel('Sales')
plt.title('Sales of each restaurant for each year')
plt.xticks(index + 1.5 * bar_width, ['Starbucks', 'RV\'s Kitchen', 'McDonald\'s', 'Dominos Pizza', 'La pinoz Pizza'])
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector", vector1 * vector2)

print("\nDot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
